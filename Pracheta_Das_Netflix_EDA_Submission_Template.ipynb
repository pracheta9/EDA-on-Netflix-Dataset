## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#Filling missing values:
df['director'].fillna('Not Known', inplace=True)
df['cast'].fillna('Not Known', inplace=True)
df['country'].fillna('Not Known', inplace=True)
df['rating'].fillna('Not Known', inplace=True)

df.isnull().sum()
````

````python
#Converting date_added to datetime format:
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')

#Extracting year from date-time:
df['year_added']=df['date_added'].dt.year
df

#Filling missing values in date_added and year_added with mode of year_added:
df['date_added'].fillna(df['year_added'].mode()[0],inplace=True)
df['year_added'].fillna(df['year_added'].mode()[0],inplace=True)

df.isnull().sum()
````

````python
#Extracting duration_time and duartion_type from duration:
df[['duration_time','duration_type']]=df['duration'].str.extract(r'(\d+)\s*(\w+)',expand=True)

#Converting duration time into numeric:
df['duration_time']=pd.to_numeric(df['duration_time'], errors='coerce')

df[['duration_time','duration_type']]
````

#### Chart - 2: Top 10 genres around the world:


````python
# Chart - 2 visualization code:

#Seperating all genres from the genre column for all movies/shows
genres = df['listed_in'].str.split(', ').explode()
top_genre=genres.value_counts().nlargest(10)
print(top_genre)

plt.figure(figsize=(10, 6))
ax=sns.barplot(y=top_genre.index,x=top_genre.values, palette='viridis')
plt.title("Top genres around the world")
plt.xlabel("Frequency")
plt.ylabel("Genre")

#For values:
for i, v in enumerate(top_genre.values):
    ax.text(v + 1, i, str(v), color='black', va='center')

plt.show()
````

#### Chart - 10

````python
# Chart - 10 visualization code:

#Filtering for 3 popular countries
popular_countries = ['United States', 'India', 'United Kingdom']
df = df[df['country'].isin(popular_countries)]

#Removing rows where cast is 'Unknown'
df = df[df['cast'] != 'Not Known']

#Figures for pie charts
plt.figure(figsize=(18,10))

#1. United States
us_casts = df[df['country'] == 'United States']['cast'].str.split(', ').explode()
us_top5 = us_casts.value_counts().head(5)

plt.subplot(2,2, 1)
plt.pie(us_top5.values, labels=us_top5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Actors - United States')

#2. India
india_casts = df[df['country'] == 'India']['cast'].str.split(', ').explode()
india_top5 = india_casts.value_counts().head(5)

plt.subplot(2,2, 2)
plt.pie(india_top5.values, labels=india_top5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Actors - India')

#3. United Kingdom
uk_casts = df[df['country'] == 'United Kingdom']['cast'].str.split(', ').explode()
uk_top5 = uk_casts.value_counts().head(5)

plt.subplot(2,2, 3)
plt.pie(uk_top5.values, labels=uk_top5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Actors - United Kingdom')

#final
plt.suptitle("Top 5 actors in each of the top 3 popular countries")
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Pie charts are useful to show proportional dominance among a small number of categories. Since the focus is on the top 5 actors per country, the pie format makes it visually engaging to compare their relative contributions.

Also, breaking it down into three subplots by country helps identify localization trends in casting across major markets.

##### 2. What is/are the insight(s) found from the chart?

Anupam Khere and Adam Sandler are the top actors un US and India, while the tope 5 actors in UK are equally popular

Are there any insights that lead to negative growth? Justify with specific reason.
